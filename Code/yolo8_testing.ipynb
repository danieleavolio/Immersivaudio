# Yolo8 testing

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.pt')  # load an official model
```

```python
import os
# Prediction
# Predict a single image

filename = 'room.mp4'


# read images from images folder
images = os.listdir(f'Scripts/{filename.split(".")[0]}/{filename}')

# sort images by number
# ['1.jpg', '10.jpg', '11.jpg', '2.jpg', '3.jpg', '4.jpg', '5.jpg', '6.jpg', '7.jpg', '8.jpg', '9.jpg']

images.sort(key=lambda x: int(x.split('.')[0]))


res = []

for image in images:
    results = model(f"Scripts/{filename.split('.')[0]}/{filename}/{image}", conf=0.20)
    res.append(results)

print(res)
```
