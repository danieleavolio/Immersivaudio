# Yolo8 testing

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.pt')  # load an official model
```

```python
import os
# Prediction
# Predict a single image

# read images from images folder
images = os.listdir('Scripts/cat/cat.mp4')

res = []

for image in images:
    results = model("Scripts/cat/cat.mp4/" + image)
    res.append(results)

print(res)
```

```python
for ix, r in enumerate(res):
    # save the predictions to results dir with the name
    r[0].save('results/' + str(ix) + '.jpg')
```
