```python
import matplotlib.pyplot as plt
```

```python
# graph containing the equations
# y = 1/2 * x
# y = 10 * log(x,10)
# y = 5 * log(x,10)
# y = 15 * log(x,10)

import math

max = 100
x = range(1, max)
y1 = [((i * 2)) for i in x]
y2 = [(10 * (math.log(i,10))) for i in x]
y3 = [(5 * (math.log(i,10)) )for i in x]
y4 = [(15 * (math.log(i,10))) for i in x]

plt.plot(x, y1, label='s = 2s')
plt.plot(x, y4, label='s = 15 * log(s)')
plt.plot(x, y2, label='s = 10 * log(s)')
plt.plot(x, y3, label='s =   5 * log(s)')


# integer values only
plt.xticks(range(0, max + 1, 20))
plt.yticks(range(0, max + 1, 20))
plt.axis([1, max, 1, max])
plt.grid(True)  
plt.xlabel('input seconds')
plt.ylabel('extracted frames')
plt.legend()
plt.show()
```
